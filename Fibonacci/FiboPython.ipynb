```python
import matplotlib.pyplot as plt
import datetime
```

```python
def fib(n):
    timeini = datetime.datetime.now()
 
    x1 = 1
    x2 = 1
 
    for i in range(n):
        tmp = x2
        x2 += x1
        x1 = tmp
 
    timefin = datetime.datetime.now()
 
    return (timefin-timeini).microseconds
```

```python
X = []
Y = []
for i in range(1000):
    X.append(i*10)
    Y.append(fib(i*10))
```

```python
plt.scatter(X,Y)
plt.xlabel('Fibonacci')
plt.ylabel('Tiempo(milisegundos)')
plt.show()
```
